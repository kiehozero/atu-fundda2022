# Exercise 7a - Replacing Characters

Write a Python function to remove all non-alphanumeric characters from a string.

```python
import urllib.request
import re
```

```python
# sample string from Wikipedia article in README
string = 'F♯ A♯ ∞ (pronounced "F-sharp, A-sharp, Infinity") is the debut studio album by Canadian post-rock band Godspeed You! Black Emperor.'

# a function to remove all alphanumeric characters
def non_w(string):
    pattern = '\w'
    new_string = re.sub(pattern, '', string)
    print(new_string)

# run the function
non_w(string)
```

```python
# code adapted from Ian's version

url = r'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

# list comprehension that loops through the text and converts each item into text
iris = [line.decode('utf-8').strip() for line in urllib.request.urlopen(url)]
iris
```

```python
# regex here looks for two digits seperated by a point (note the escape slash), four times, then looks for a 
# final text value. Crucially, only the part after the dash is in brackets, so this is the only part passed by
# the sub function to the new variable.
strip_iris = re.compile(r'([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),Iris-([a-z]+)')

# loops through each line in the file and reverses the order in which items are displayed
strip_iris = [strip_iris.sub(r'\5,\4,\3,\2,\1', line) for line in iris if line]
strip_iris
```

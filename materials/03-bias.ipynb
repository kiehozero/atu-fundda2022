<h2 style="color: rgb(0, 91, 94);">Statistical Bias</h2>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

Two of the common calculations you will find in the statistics literature are mean and standard deviation.

The mean is straight-forward - it is the usual calculation that people call the average.

You take all of the numbers you have, add them up, and then divide by the number of them.

<h3 style="color: rgb(0, 91, 94);">Mean</h3>

<i>SS - note that the video is titled "MEAN IS UNBIASED".</i>

Suppose you take a sample of values from a larger population of values.

```python
# Generate a sample of 1000 values from a normal distribution.
x = np.random.normal(10.0, 1.0, 1000)
# SS - in the normal distribution, a standard deviation of 1 means that 99% of the results will be with in 3
# of the mean (10). You will often see the mean expressed as a variable called 'mu', while the standard deviation 
# will be expressed as 'sigma': https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
x
```


If you take the mean of the sample, it is a good estimate of the population average.

```python
# SS - We expect the mean of the sample to be close to the mean of the population. This will be close to the 
# mean given above, but rarely the same
x.mean()
```

What do mean by a good estimate?

To investigate, let us take lots of samples.

```python
# Let's run a simulation of taking 1000 samples of size 1000.
samples = np.random.normal(10.0, 1.0, (1000, 1000))
# this is going to produce an array that is one thousand columns wide and one thousand rows long
samples
```

```python
# Get the mean of the first sample.
samples[0].mean()
```

```python
# Calculate the mean of each sample produced.
sample_means = samples.mean(axis=1)
# SS - the axis parameter defines how to bin the process, so an axis of zero would produce the mean per column, while 
# an axis of 1 produces 1 mean per row. The default of blank just produces a mean for the entire 1000 x 1000 array: 
# https://www.geeksforgeeks.org/numpy-mean-in-python/
sample_means
```

Let's plot the means in a histogram.

<i>SS - it is useful to remember the process above. A thousand normal distribution samples of a thousand numbers were generated, and then the mean of each was calculated. So the sample_means variable actually contains the mean of one thousand normally-distributed datasets, rather than one thousand normally-distributed random numbers, so even more of these numbers will actually be very close to the mean number original provided. Calculating the mean of the means of a distribution should be a good estimate of what that generated mean actually is.</i>

```python
plt.hist(sample_means);
```

<h3 style="color: rgb(0, 91, 94);">Standard Deviation</h3>

<i>SS - note that the video is titled "STANDARD DEVIATION IS BIASED". A bar chart is used for discrete numbers, generally things that can be counted, while histograms are used for numbers that are continuous, or cannot be counted. While the histogram may look like it has bars, these are in fact bins and merely represent an arbitrary point at which the items are separated, rather than them being a different measure.<i>

The standard deviation is a different story.

First of all, the calculation is not as familiar.

It is designed to give a measure of how far the numbers are away from the mean in general.

The need for a such a measure is seen in the following example of calculating the mean of two sets of numbers.

```python
# A list of numbers - four small and one big.
numbers1 = np.array([1, 1, 1, 1, 10])
# Their mean.
np.mean(numbers1)
```

```python
# A list of numbers - all close to each other.
numbers2 = np.array([2, 2, 3, 3, 4])
# Their mean.
np.mean(numbers2)
```

The example illustrates a common issue.

The mean on its own does is limited as a summary of the data points.

That is why we use the standard deviation - it gives us a measure of the spread.

Let's see how it is calculated.

First we'll generate a sample.

<i>SS - this is a step-by-step guide to calculating the standard deviation.<i/>

```python
# Generate a sample of values - note we can see the population standard deviation.
x = np.random.normal(10.0, 1.0, 1000)
x
```

Now, let us calculate the mean and investigate it.

```python
# Calculate the mean.
x_mean = x.mean()

# Subtract the mean from each of the values.
zeroed = x - x_mean

print(x_mean)

zeroed
```

<i>SS - note how these are all written in exponential notation because the values are small. Each of the numbers above is the a value in the original array, but minus the mean of that array as a whole. Remember that the final two digits in exponential notation denote how far the decimal point moves to the right, so the a value which is 4.68275493e-01 is in fact 0.468</i>

```python
# What do you think the mean of zeroed is?
zeroed.mean()
```

```python
# This will give us a better view of it - correct to four decimal places.
print(f'{zeroed.mean():0.4f}')
```

So, subtracting the mean of the sample results in the mean being zero.

The standard deviation is an adjustment to the above calculation.

The goal is to summarise the zeroed array without losing information about the distance of each point from the mean.

Let's see if we can come up with a plot of the idea.

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the zeroed array, each value spaced out evenly along the x axis.
# Note the x axis is just the position of the value in the zeroed array.
ax.plot(range(len(zeroed)), zeroed, 'k.')

# Plot the y=0 line.
ax.axhline(y=0.0, color='grey', linestyle='-');
```

One idea is to take the average (vertical) distance each point is from the mean, zero.

We need to be careful - there are negative and positive values here.

By definition, they sum to zero.

SS - the idea of the normal distribution is to have the same number of points either side of the mean value. Think of the above plot, turned on it's side. Imagine some number of bins between 3 and -3 that count each value and plot them, it would create the familiar normal distribution curve, it can be seen clearly that many of the points are grouped around the centre line of 0, with progressively less further out.

```python
# Sum the array.
print(f'{zeroed.sum():0.4f}')
```

We can try instead taking the absolute value.

This makes sense, because we expect distances to be positive.

<i>SS - absolute value just converts any negative number into a positive number, so -1.1 becomes 1.1. The items will also still be the same distance to zero, but the negatives will no longer cancel out the positives.</i>

```python
# Absolute values.
np.abs(zeroed)
```

```python
# Average absolute value.
np.mean(np.abs(zeroed))
```

While this is a reasonable measure of spread, it is not typically the one used.

For a discussion, see this Cross Validated post: https://stats.stackexchange.com/q/118

Instead we usually square the values.

Note squaring a number gives a positive value.

It is also somewhat easier to work with analytically.

```python
# Square the values.
np.square(zeroed)
```

Note that when you square numbers, bigger values get bigger.

Larger deviations from the mean will contribute relatively more to the standard deviation.

That is not necessarily a bad thing, it is just something to note.

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the squared zeroed array, each value spaced out evenly along the x axis.
# Note the x axis is just the position of the value in the zeroed array.
ax.plot(range(len(zeroed)), np.square(zeroed), color='green', marker='.', linestyle='none')

# Plot the zeroed array, each value spaced out evenly along the x axis.
# Note the x axis is just the position of the value in the zeroed array.
ax.plot(range(len(zeroed)), zeroed, 'k.')

# Plot the y=0 line.
ax.axhline(y=0.0, color='grey', linestyle='-');
```

<i>SS - in looking at the plot above, you can actually trace the green dots back down the y-axis to their original black dots, as both plots are using the same `range(len(zeroed))` function to plot points on the x-axis.

```python
# Calculate the average squared result.
np.mean(np.square(zeroed))
```

Now, because we have squared the original values, we often take the square root of the answer.

```python
# Calculate the square root of the average squared result.
# SS - this square process is just known as the variance; once it is square-rooted, it is known as the standard deviation.
np.sqrt(np.mean(np.square(zeroed)))
```

So, here is the full calculation.

```python
# The full calculation using the original array.
np.sqrt(np.mean(np.square(x - np.mean(x))))
```

This common calculation is built into numpy.

## Note this is very close (by design) to the second parameter we sent to `np.random.normal`.

<i>SS - it is very close because random.normal takes that initial parameter and then randomises data around it, it is possible for them to be the exactly same, but no more like that any other similar amount.</i>

```python
# Note that the function is built into numpy.
x.std()
```

<h2 style="color: rgb(0, 91, 94);">Bessel's Correction</h2>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

The above calculation of the standard deviation has one flaw.

If you calculate the standard deviation of a sample, it is a biased estimator for the standard deviation of the population.

<h3 style="color: rgb(0, 91, 94);">Biased Estimators</h3>

The issue is apparent when we repeatedly draw samples from a population.

If we use the above calculation on each sample, we will systematically under-estimate the standard deviation.

Let us see if we can see this in a plot.

We will use small samples as the effect is clearer.

```python
# Create 100000 samples of size 5 - standard deviation is 2.0 and the mean is 0.0.
samples = np.random.normal(0.0, 2.0, (100000, 5))
# a feature of the normal distribution is that if you multiple the given standard deviation (in this case 2.0) 
# by 3, you'll capture 99% of the items in the total range. So 99% of the items will be between -6 and 6.
samples
```

```python
# Calculate standard deviation without correction.
# SS - the axis parameter defines how to bin the process, so an axis of zero would produce the mean per column, while 
# an axis of 1 produces 1 mean per row. The default of blank just produces a mean for the entire 1000 x 1000 array: 
# https://www.geeksforgeeks.org/numpy-mean-in-python/
# in this case the axis parameter reduces the total number of samples from 500,000 (an array of 5 random items, repeated
# 100,000 times, down to an average of each of those repetitions).
stdevs = samples.std(axis=1)
stdevs
```

```python
# View a histogram - hopefully we can see the estimate is too small.
fig, ax = plt.subplots(figsize=(12, 6))

# Plot histogram.
plt.hist(stdevs, bins=100)

# Draw a vertical line where the actual standard deviation is.
plt.axvline(x=2.0, color='red');
```

It seems clear that the tip of the curve is below the actual value.

<i>SS - remember that this is a normal distribution, and yet it does not look like a normal distribution. Note the point above about under-estimation; the standard deviation is actually larger than the modal average, and the decline on the higher side of the range is not as steep as the incline at the lower end. Note also how in the earlier demonstrations, the final standard deviation was very close to the original standard deviation parameter given in `np.random.normal`, often within 0.01.

<h3 style="color: rgb(0, 91, 94);">The Correction</h3>

Bessel's correction actually applies to the variance.

The variance of a sample is the square of the standard deviation.

It is what we get is we do not apply the `np.sqrt` function in the standard deviation calculation.

The correction is to mutliply the calculation be $\frac{n}{n-1}$.

```python
# Uncorrected variance.
np.mean(np.square(x - np.mean(x)))
```

```python
# Corrected variance.
np.mean(np.square(x - np.mean(x))) * (len(x) / (len(x) - 1.0))
```

The correction can be applied to the standard deviation directly, but will still lead to a somewhat biased estimator.
